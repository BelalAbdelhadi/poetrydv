```python
#Importing needed libraries
import pandas as pd
import numpy as np
import random
import datetime
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns
import os
from matplotlib import cm
from calendar import monthrange
pd.options.mode.chained_assignment = None
%matplotlib inline
```

```python
#Getting the new filenames corresponding to weeks each sheet represents.
filenames = []
for file in os.listdir('2016'):
    df = pd.read_excel('2016/{}'.format(file), engine='openpyxl')
    new_filename = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].strip()
    filenames.append(f"{new_filename}.xlsx")
```

```python
#Function for determining the month based on the week
def consider_first(d1: datetime.date, d2: datetime.date) -> bool:
    number_of_days = monthrange(d1.year, d1.month)[1]
    return number_of_days - d1.day + 1 > d2.day
```

```python
#Storing each sheet into a different dataframe while performing:
#1. Basic Cleaning
#2. Adding a week start column to correspond to the week each book's info was collected in
#3. Adding a month column to correspond to the month (will be needed in time series analysis)
Poetry_2016 = []
for file in os.listdir('2016'):
    df = pd.read_excel('2016/{}'.format(file), engine = 'openpyxl')
    week_start = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].split('-')[0].strip()
    week_end = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].split('-')[1].strip()
    df = pd.read_excel('2016/{}'.format(file), engine = 'openpyxl', skiprows=8, skipfooter=2)
    df.dropna(axis=1, inplace=True)
    df.rename(columns = lambda x:x.strip(), inplace=True)
    week_start = datetime.datetime.strptime(week_start, "%b %d %Y")
    week_end = datetime.datetime.strptime(week_end, "%b %d %Y")
    df['Week Start'] = week_start
    df['Week End'] = week_end
    if week_start.month != week_end.month:
        _consider_first = consider_first(week_start, week_end)
        if _consider_first:
            df['Month'] = df['Week Start'].dt.month
        else:
            df['Month'] = df['Week End'].dt.month
    else: 
        df['Month'] = df['Week Start'].dt.month
    Poetry_2016.append(df)
```

```python
#Concatenating all the dataframes into one and sorting it by date
Poetry_2016_df = pd.concat(Poetry_2016)
Poetry_2016_df.sort_values(['Week Start', 'TW Rank'], inplace=True)
```

```python
#Manually adding the info for the missing book
manual2_df = Poetry_2016_df[Poetry_2016_df['ISBN'] == 9781682410189]
manual_df = manual2_df.copy(deep=True)
manual_df['Title'] = 'Whiskey, Words, And a shovel'
manual_df['Author'] = 'R.H Sin.'
manual_df['Publisher'] = 'Andrews & McMeel'
manual_df['Imprint'] = 'Andrews Mcmeel Publishing'
manual_df['Publication Date'] = datetime.datetime(2015, 12, 25, 0, 0, 0)
manual_df['Price'] = 14.99 
manual_df['Format'] = 'Trade Paperback'
manual_df['Bisac Level 1'] = 'POETRY'
manual_df['Bisac Level 2'] = 'POETRY  / American'
Poetry_2016_df[Poetry_2016_df['ISBN'] == 9781682410189] = manual_df
```

### Market Shares

```python
author_shares = Poetry_2020_df.groupby('Author').sum().drop(['ISBN', '#Weeks on List', 'TW Rank', 'YTD Sales', 'LW Rank', 'Price', 'Month'], axis=1)
author_shares.rename(columns = {'TW Sales': 'Sales'}, inplace=True)
author_shares.sort_values('Sales', inplace=True, ascending = False)
authors_2020_top10 = author_shares.iloc[0:10]
rest = sum(author_shares.iloc[10:-1].values)
others = pd.Series(rest, index=['Others'])
authors_2020_top10['Author'] = authors_2020_top10.index
authors_2020_top10 = authors_2020_top10.append(others, ignore_index=True)
authors_2020_top10.Sales.iloc[10] = rest
authors_2020_top10.Author.iloc[10] = 'Others'
authors_2020_top10.drop(['Others'], axis=1, inplace=True)
authors_2020_top10
```

```python
colors = random.choices(list(mcolors.CSS4_COLORS.values()),k = 11)
plt.figure(figsize=[10,10])
plt.pie(authors_2020_top10.Sales, labels = authors_2020_top10.Author, autopct='%1.1f%%', startangle=90, rotatelabels=True, colors=colors);
```

```python
Poetry_2020_df['Net Gross'] = Poetry_2020_df['TW Sales']* Poetry_2020_df['Price']
```

```python
author_shares_gross = Poetry_2020_df.groupby('Author').sum().drop(['ISBN', '#Weeks on List', 'TW Rank', 'YTD Sales', 'LW Rank', 'Price', 'Month', 'TW Sales'], axis=1)
author_shares_gross.sort_values('Net Gross', inplace=True, ascending = False)
authors_2020_top10 = author_shares_gross.iloc[0:10]
rest = sum(author_shares_gross.iloc[10:-1].values)
others = pd.Series(rest, index=['Others'])
authors_2020_top10['Author'] = authors_2020_top10.index
authors_2020_top10 = authors_2020_top10.append(others, ignore_index=True)
authors_2020_top10['Net Gross'].iloc[10] = rest
authors_2020_top10.Author.iloc[10] = 'Others'
authors_2020_top10.drop(['Others'], axis=1, inplace=True)
authors_2020_top10
```

```python
for file in os.listdir('2021'):
    df = pd.read_excel(f'2021/{file}', engine = 'openpyxl')
    new_filename = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].strip()
    if new_filename != file:
        os.rename(f"2021/{file}", f"2021/{new_filename}.xlsx")
    else:
        continue
```

```python
#Storing each sheet into a different dataframe while performing:
#1. Basic Cleaning
#2. Adding a week start column to correspond to the week each book's info was collected in
#3. Adding a month column to correspond to the month (will be needed in time series analysis)
Poetry_2021 = []
for file in os.listdir('2021'):
    df = pd.read_excel('2021/{}'.format(file), engine = 'openpyxl')
    week_start = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].split('-')[0].strip()
    week_end = df['Unnamed: 0'][2].split(':')[2].split('\n')[0].split('-')[1].strip()
    df = pd.read_excel('2021/{}'.format(file), engine = 'openpyxl', skiprows=8, skipfooter=2)
    df.dropna(axis=1, inplace=True)
    df.rename(columns = lambda x:x.strip(), inplace=True)
    week_start = datetime.datetime.strptime(week_start, "%b %d %Y")
    week_end = datetime.datetime.strptime(week_end, "%b %d %Y")
    df['Week Start'] = week_start
    df['Week End'] = week_end
    if week_start.month != week_end.month:
        _consider_first = consider_first(week_start, week_end)
        if _consider_first:
            df['Month'] = df['Week Start'].dt.month
        else:
            df['Month'] = df['Week End'].dt.month
    else: 
        df['Month'] = df['Week Start'].dt.month
    Poetry_2021.append(df)
#Concatenating all the dataframes into one and sorting it by date
Poetry_2021_df = pd.concat(Poetry_2021)
Poetry_2021_df.sort_values(['Week Start', 'TW Rank'], inplace=True)
Poetry_2021_df
```
